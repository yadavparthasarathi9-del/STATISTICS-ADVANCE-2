#**1**


A statistical technique for determining whether a claim (hypothesis) about a population has sufficient evidence in a sample of data to support or refute it is called hypothesis testing.

 It includes:

 The default assumption (no effect or no difference) is known as the null hypothesis (H‚ÇÄ).

 What you wish to test or prove is known as the alternative hypothesis (H‚ÇÅ).

 The test statistic, also known as the p-value, is used to determine whether to reject H‚ÇÄ.

 ‚úÖ  If the p-value is less than the significance level (Œ±), then reject H‚ÇÄ.  Failure to reject H‚ÇÄ if p-value ‚â• Œ±

#**2**
The null hypothesis (H‚ÇÄ) states that there is no effect or no difference ‚Äî it‚Äôs the default assumption.
The alternative hypothesis (H‚ÇÅ or Ha) states that there is an effect or a difference ‚Äî it‚Äôs what you want to prove.

üëâ Example:

H‚ÇÄ: The mean weight = 70 kg

H‚ÇÅ: The mean weight ‚â† 70 kg

#**3**
The significance level (Œ±) is the threshold for deciding whether to reject the null hypothesis.

It represents the probability of making a Type I error ‚Äî rejecting H‚ÇÄ when it‚Äôs actually true.

Common values: 0.05, 0.01, or 0.10.

üîπ If p-value < Œ±: Reject H‚ÇÄ (evidence supports H‚ÇÅ)
üîπ If p-value ‚â• Œ±: Fail to reject H‚ÇÄ (not enough evidence against H‚ÇÄ)

So, Œ± controls how strict the test is in detecting a true effect.

#**4**


Type I Error (False Positive):
Rejecting the null hypothesis when it is actually true.
üëâ Example: Concluding a new drug works when it actually doesn‚Äôt.

Type II Error (False Negative):
Failing to reject the null hypothesis when it is actually false.
üëâ Example: Concluding a new drug doesn‚Äôt work when it actually does.

#**5**
**Z-test:**

* Used when **sample size is large (n > 30)**
* **Population standard deviation (œÉ)** is known
* Follows a **normal distribution**

**T-test:**

* Used when **sample size is small (n ‚â§ 30)**
* **Population standard deviation (œÉ)** is unknown (use sample SD instead)
* Follows a **t-distribution**




```python
#6
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10     # number of trials
p = 0.5    # probability of success
size = 1000  # number of experiments

# Generate binomial data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#7
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-test statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

# Significance level
alpha = 0.05

if p_value < alpha:
    print(" Reject the null hypothesis (H‚ÇÄ). There is a significant difference.")
else:
    print(" Fail to reject the null hypothesis (H‚ÇÄ). No significant difference.")
```

```python

#8
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(0)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

# Step 2: Calculate sample statistics
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)

# Step 3: Calculate 95% confidence interval for the mean
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # two-tailed t critical value
margin_of_error = t_critical * (std_dev / np.sqrt(n))

ci_lower = mean - margin_of_error
ci_upper = mean + margin_of_error

print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot the data
plt.figure(figsize=(8, 5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"Upper 95% CI = {ci_upper:.2f}")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```

```python
#9
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores and plot histogram
def z_score_analysis(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Display results
    print(f"Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}")
    print("First 5 Z-scores:", z_scores[:5])

    # Plot histogram of Z-scores
    plt.figure(figsize=(8,5))
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.legend()
    plt.show()

    return z_scores

# Example dataset
data = np.random.normal(loc=50, scale=5, size=100)

# Call the function
z_scores = z_score_analysis(data)
```

Z Score reprsents that

They measure how many standard deviations each data point is from the mean.

A Z = 0 means the value is exactly at the mean.

Positive Z ‚Üí above the mean; Negative Z ‚Üí below the mean.

Example: Z = 2 means the value is 2 standard deviations above the mean.
